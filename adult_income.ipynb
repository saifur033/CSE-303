```python
import pandas as pd

# Load the CSV file
file_path = '/content/drive/MyDrive/CSE 303/lab 07/adult_income.csv'
 # Corrected file path
data = pd.read_csv('/content/drive/MyDrive/CSE 303/lab 07/adult_income.csv')

# Display the first 5 and last 5 rows of the dataset
print("First 5 rows:\n", data.head())
print("\nLast 5 rows:\n", data.tail())
```

**Scatter plot: Age vs Hours-per-week colored by income**

Individuals with incomes above 50K tend to work longer hours across a wide range of ages.

Clusters of individuals with <=50K income show a more concentrated working range (around 40 hours/week).

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the style for plots
sns.set(style="whitegrid")

# Scatter Plot: Age vs Hours-per-week colored by income
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='age', y='hours-per-week', hue='income', alpha=0.7)
plt.title(' Age vs Hours-per-week colored by Income')
plt.show()
```

**Line plot:Mean educational-num by age**

Educational attainment generally increases in younger age groups up to a certain point .

A gradual decline or stabilization in older ages indicates education levels are relatively stable after certain career stages.


```python
mean_edu_by_age = data.groupby('age')['educational-num'].mean()
plt.figure(figsize=(10, 6))
mean_edu_by_age.plot(kind='line', color='blue')
plt.title(' Mean Educational-num by Age')
plt.xlabel('Age')
plt.ylabel('Mean Educational-num')
plt.show()
```

**Bar plot: Count of income categories**

A significantly larger portion of the population earns <=50K, indicating income inequality.

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='income', palette='Set2')
plt.title('Income Category Count')
plt.show()
```

**Pie chart: Proportion of workclasses**

Most individuals belong to the "Private" workclass, while smaller segments are distributed across other categories like "Self-emp-inc" and "Local-gov".

```python
workclass_counts = data['workclass'].value_counts()
plt.figure(figsize=(20, 20))
workclass_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colormap='viridis', legend=True)
plt.title('Pie Chart: Workclass Distribution')
plt.ylabel('')
plt.show()

```

**Histogram: Distribution of Hours-per-week**

A sharp peak around 40 hours/week shows this is a common working norm.

A smaller number of individuals work extremely low or high weekly hours

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='hours-per-week', kde=True, color='green', bins=30)
plt.title('Hours-per-week Distribution')
plt.show()
```

**Boxplot: Capital-gain distribution by income**

Individuals with >50K income tend to have higher capital gains, though outliers exist.

Most individuals with <=50K income have negligible or no capital gains.


```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=data, x='income', y='capital-gain', palette='coolwarm')
plt.title('B Capital-gain by Income')
plt.ylim(0, 10000)  # Limit for better visibility
plt.show()
```

**Count Plot: Education levelst**
The most common education levels are HS-grad and Some-college," reflecting a broad middle-education demographic.

```python
plt.figure(figsize=(14, 6))
sns.countplot(data=data, y='education', order=data['education'].value_counts().index, palette='mako')
plt.title(' Education Levels')
plt.show()
```
