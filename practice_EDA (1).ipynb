**Scatter Plot :**
This plot visualizes the relationship between the number of confirmed cases and deaths across regions



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the data:
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Confirmed', y='Deaths', hue='WHO Region', data=df, palette='Set2')  # Change 'data=data' to 'data=df'
plt.title('Scatter Plot: Confirmed Cases vs Deaths')
plt.xlabel('Confirmed Cases')
plt.ylabel('Deaths')
plt.legend(title='WHO Region')
plt.show()
```

**Line Plott:** Weekly Change in Confirmed Cases

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' or a similarly named DataFrame contains the data:
# Replace 'df' with the actual name of your DataFrame if it's different.
data = df  # Assign your DataFrame to the variable 'data'

plt.figure(figsize=(8, 6))
sns.lineplot(x=data.index, y='1 week change', hue='WHO Region', data=data, palette='Set1', marker="o")
plt.title('Line Plot: Weekly Change in Confirmed Cases')
plt.xlabel('Countries (Index)')
plt.ylabel('Weekly Change in Cases')
plt.legend(title='WHO Region')
plt.show()
```

** Bar Plot**

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='WHO Region', y='Deaths', data=data, palette='viridis', errorbar=None)
plt.title('Bar Plot: Total Deaths by WHO Region')
plt.xticks(rotation=45)
plt.ylabel('Total Deaths')
plt.show()
```

**Pie Chart**

```python
plt.figure(figsize=(8, 6))
region_counts = data['WHO Region'].value_counts()
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Pie Chart: Distribution of WHO Regions')
plt.show()
```

**Histogram**

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Confirmed'], bins=30, kde=True, color='blue')
plt.title('Histogram: Distribution of Confirmed Cases')
plt.xlabel('Confirmed Cases')
plt.ylabel('Frequency')
plt.show()
```

**Boxplot**

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='WHO Region', y='New cases', data=data, palette='coolwarm')
plt.title('Boxplot: New Cases by WHO Region')
plt.xticks(rotation=45)
plt.ylabel('New Cases')
plt.show()
```

**Count Plot**

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='WHO Region', data=data, palette='muted')
plt.title('Count Plot: Number of Countries per WHO Region')
plt.xticks(rotation=45)
plt.ylabel('Count')
plt.show()
```
